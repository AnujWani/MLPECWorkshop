```python
import numpy as np
```

```python
file_path = 'code.txt'
```

```python
def get_Text( file_path ):
    with open( file_path, encoding='utf-8' ) as f:
        return f.read().lower()
```

```python
text = get_Text(file_path)
```

```python
def gen_table( text, k =4 ):
    T = {}
    for i in range( len(text) - k ):
        X = text[ i:i+k ]
        Y = text[i+k]
        
        if T.get( X ) is None:
            T[ X ] = {}
            T[X][ Y ] = 1
        elif T[X].get(Y) is None:
            T[X][Y] = 1
        else:
            T[X][Y] += 1
            
    return T
        
        
```

```python
test = "My name is Suransh Chopra"
```

```python
out = gen_table( text )
```

```python
def get_prob( T, k = 4 ):
    for i in list( T.keys() ):
        s = float( sum( list( T[i].values() ) ) )
        for j in list( T[i].keys() ):
            T[i][j] = float( T[i][j] ) / s
    return T
```
